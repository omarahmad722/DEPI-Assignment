### 1. Question: Merge and Sort Two Lists

**Description:** Given two lists of integers, merge them into one list and sort the result in ascending order. Do not use Python's built-in sort() function or other sorting functions.




```python
'''
# Example Input:
list1 = [3, 1, 4]
list2 = [2, 5, 0]
# Example Output:
[0, 1, 2, 3, 4, 5]
'''
```

```python
list1=list(map(int,input("Enter list1:").split(" ")))
list2=list(map(int,input("Enter list2:").split(" ")))
x=list1+list2
print("before sort:",x)
n=len(x)
for i in range(len(x)):
    for j in range(i+1,len(x)):
        if x[i]>x[j]:
            y=x[i]
            x[i]=x[j]
            x[j]=y
print("after sort:",x)
```
